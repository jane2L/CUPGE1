```julia
using Images, ImageIO, ImageView
using ColorTypes
using ImageTransformations
using GLMakie
using LinearAlgebra

function grey_transform(img)
    h, w = size(img)  # hauteur et largeur
    img_grey = similar(img)  # crée une nouvelle image de mêmes dimensions
    G = [0.2126 0.7152 0.0722 ; 0.2126 0.7152 0.0722 ; 0.2126 0.7152 0.0722]  # Matrice de transformation en niveaux de gris

    for i in 1:h
        for j in 1:w
            px = img[i, j]
            r = Float64(red(px))
            g = Float64(green(px))
            b = Float64(blue(px))
            
            # Calcul du niveau de gris
            gval = G*[r ;g; b]

            # Remplit chaque canal R, G, B avec le même niveau de gris
            img_grey[i, j] = RGB{Float32}(gval[1], gval[2], gval[3])
        end
    end
    return img_grey
end


function sepia_transform(img)
    h, w = size(img)  # hauteur et largeur
    img_sepia = similar(img)  # crée une nouvelle image de mêmes dimensions
    S = [0.393 0.769 0.189 ; 0.349 0.686 0.168 ; 0.272 0.534 0.131] # Matrice de transformation en sepia

    for i in 1:h
        for j in 1:w
            px = img[i, j]
            r = Float64(red(px))
            g = Float64(green(px))
            b = Float64(blue(px))
            
            # Calcul du niveau de gris
            sval = S*[r ;g ;b]
            
            # Remplit chaque canal R, G, B avec la teinte de sepia
            img_sepia[i, j] = RGB{Float32}(max(0,min(sval[1],1)), max(0,min(sval[2],1)), max(0,min(sval[3],1)))
        end
    end
    return img_sepia
end

function pixel_list(img) # retourne une matrice hauteur*largeur x 3 contenant les pixes en ligne et les valeurs r,g,b en colonne.
    h, w = size(img)  # hauteur et largeur
    pixels = zeros(Float32, h * w, 3)
    for i in 1:h
        for j in 1:w
            px = img[i, j]
            r = Float32(red(px))
            g = Float32(green(px))
            b = Float32(blue(px))
            idx = (i - 1) * h + j
            pixels[idx, :] .= [r, g, b]
        end
    end
    return pixels
end

function TO_transform(img)
    h, w = size(img)  # hauteur et largeur
    img_sepia = similar(img)  # crée une nouvelle image de mêmes dimensions
    #TO = [1.1 0.1 -0.2 ; -0.1 1.05 0.05 ; 1 1.15 -0.15] # Matrice de transformation en sepia
    TO = [0.9 0.3 -0.2; 0 0.7 0.3; 0.9 1 0.1]
    for i in 1:h
        for j in 1:w
            px = img[i, j]
            r = Float64(red(px))
            g = Float64(green(px))
            b = Float64(blue(px))
            
            # Calcul du niveau de gris
            sval = TO*[r ;g ;b]
            # Remplit chaque canal R, G, B avec la teinte de sepia
            img_sepia[i, j] = RGB{Float32}(max(0,min(sval[1],1)), max(0,min(sval[2],1)), max(0,min(sval[3],1)))
        end
    end
    return img_sepia
end

function linear_transform(img,T)
    h, w = size(img)  # hauteur et largeur
    img_ = similar(img)  # crée une nouvelle image de mêmes dimensions
    #TO = [1.1 0.1 -0.2 ; -0.1 1.05 0.05 ; 1 1.15 -0.15] # Matrice de transformation en sepia
    TO = [0.9 0.3 -0.2; 0 0.7 0.3; 0.9 1 0.1]
    for i in 1:h
        for j in 1:w
            px = img[i, j]
            r = Float64(red(px))
            g = Float64(green(px))
            b = Float64(blue(px))
            
            # Calcul du niveau de gris
            tval = T*[r ;g ;b]
            # Remplit chaque canal R, G, B avec la teinte de sepia
            img_[i, j] = RGB{Float32}(max(0,min(tval[1],1)), max(0,min(tval[2],1)), max(0,min(tval[3],1)))
        end
    end
    return img_
end

# Chargement de l'image
img = load("solo.jpeg")

# Affichage des informations de l'image
println("dimensions : ", size(img))      # (hauteur, largeur)
println("type des pixels : ", eltype(img))

# Affichage de l'image
display(img)

# Coordonnées du pixel
i = 1
j = 140
println("position du pixel dans l'image : ligne ", i, " colonne ", j)

# Accès aux valeurs RGB du pixel (attention : ligne = y, colonne = x)
pixel_value = img[i, j]  # Les indices sont dans l'ordre (row, col)
ColorTypes.BGR{Float32}
r = red(pixel_value)
g = green(pixel_value)
b = blue(pixel_value)

println("valeurs (r g b) du pixel choisi : ", r, ", ", g, ", ", b)

# Création d'une image 1x1 avec ce pixel

display(RGB(r,g,b))
display(img[i,j])

# Affichage du pixel agrandi
```

```julia
pixels = pixel_list(img);
x = pixels[:,1];
y = pixels[:,2];
z = pixels[:,3];

pixels_grey = pixel_list(grey_transform(img));
xg = pixels_grey[:,1];
yg= pixels_grey[:,2];
zg = pixels_grey[:,3];

pixels_sepia = transpose((G*transpose(pixels)));
xs = pixels_sepia[:,1];
ys = pixels_sepia[:,2];
zs = pixels_sepia[:,3];

pixels_TO = transpose(TO*transpose(pixels))
xto = pixels_TO[:,1];
yto = pixels_TO[:,2];
zto = pixels_TO[:,3];
```
