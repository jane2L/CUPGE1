```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
classementPHYS = pd.read_csv("classementPHYS.csv",sep = ";",decimal = ",",dtype = np.float64)
#notes = pd.melt(notes_).rename(columns={"variable": "DS", "value": "notes"})
classementPHYS.head()
```

```python
classementCYBER = pd.read_csv("CYBERJuryEnssat.csv",sep = ";",decimal = ",")
classementCYBER.head()
```

```python
s1 = pd.Series(classementCYBER["rangCUPGE"],dtype = np.float64)
s2 = pd.Series(classementCYBER["rangAdmission"],dtype = np.float64)
s3 = pd.Series(classementCYBER["classementJuryEnssat"],dtype = np.float64)
s4 = pd.Series(classementCYBER["rangMaths"],dtype = np.float64)
s5 = pd.Series(classementCYBER["MoyAnuelle"],dtype = np.float64)
s6 = pd.Series(classementCYBER["Maths"],dtype = np.float64)
color = pd.Series(classementCYBER["couleur"],dtype = np.str_)
```

```python
s1 = pd.Series(classementPHYS["rg.semestre"])
s2 = pd.Series(classementPHYS["ordre.admission"])
s3 = pd.Series(classementPHYS["rg.admission"])
s4 = pd.Series(classementPHYS["rg.maths"])
s5 = pd.Series(classementPHYS["m.generale"])
s6 = pd.Series(classementPHYS["m.maths"])
```

```python
# Calcul des corrélations
# Calcul des corrélations
s1.corr(s2, method='kendall')
print("Correlation entre le rang de l'élève et son ordre d'admission : ", s1.corr(s2, method='kendall'))
print("Correlation entre le rang général de l'élève et son rang en maths", s1.corr(s4,method='kendall'))
print("Correlation entre la moyenne de l'élève et son classement avant admission : ", -s5.corr(s3,method = 'pearson'))
color = ['green','green','green','green','blue','blue','blue','orange','red','red','red']
```

```python
plt.scatter(s3[:],s5[:],c = color)
plt.xlabel("classement à l'admission")
plt.ylabel("moyenne générale en CUPGE")
```
