## 3. Modification de la couleur d'un pixel <a class="anchor" id="section3"></a>

### 3.1 Transformation en gris<a class="anchor" id="section21"></a>

La matrice $G$ permettant de passer d'un pixel au pixel gris correspondant est 
$$ G = \begin{pmatrix} 0.2126 & 0.7152 & 0.0722 \\ 0.2126 & 0.7152 & 0.0722 \\ 0.2126 & 0.7152 & 0.0722 \end{pmatrix}$$
On obtient le pixel gris $p'$ à partir du pixel de couleur $p$ par l'opération 
$$ p' = Gp$$

```python
# On définit la matrice G comme un tableau de dimension 2
G = np.array([[0.2126,0.7152,0.0722],[0.2126,0.7152,0.0722],[0.2126,0.7152,0.0722]])
print("G = \n", G)

# on crée un vecteur de R^3 contenant les valeurs r g b comme coordonnées.
pixel_array = np.array([r,g,b])

# multiplication matricielle. On arrondit la sortie car les valeurs r' g' b' du pixel gris doivent être entières.
pixel_array_gris = np.dot(G,pixel_array).astype(int) 
print("\n valeur r g b du pixel après transformation en gris :", pixel_array_gris[0],pixel_array_gris[1],pixel_array_gris[2] )

# Affichage du pixel transformé en gris (la taille est augmentée pour le rendre visible). 
pixel_gris = Image.new("RGB",(1,1),(pixel_array_gris[0],pixel_array_gris[1],pixel_array_gris[2]))
scaled_pixel_gris = pixel_gris.resize((100, 100), Image.NEAREST)
display(scaled_pixel_gris)
```

### 3.2 Transformation en Sepia<a class="anchor" id="section23"></a>

La matrice $S$ permettant de passer d'un pixel au pixel gris correspondant est 
$$ S = \begin{pmatrix} 0.393 & 0.769 & 0.189 \\ 0.349 & 0.686 & 0.168 \\ 0.272 & 0.534 & 0.131 \end{pmatrix}$$
On obtient le pixel sepia $p'$ à partir du pixel de couleur $p$ par l'opération 
$$ p' = Sp$$

```python
# On définit la matrice S comme un tableau de dimension 2
S = np.array([[0.393,0.769,0.189],[0.349,0.686,0.168],[0.272,0.534,0.131]])
print("S = \n", S)

# multiplication matricielle. On arrondit la sortie car les valeurs r' g' b' du pixel sepia doivent être entières.
pixel_array_sepia = np.dot(S,pixel_array).astype(int) 
print("\n valeur r g b du pixel après transformation en sepia :", pixel_array_sepia[0],pixel_array_sepia[1],pixel_array_sepia[2] )

# Affichage du pixel transformé en sepia (la taille est augmentée pour le rendre visible). 
pixel_sepia = Image.new("RGB",(1,1),(pixel_array_sepia[0],pixel_array_sepia[1],pixel_array_sepia[2]))
scaled_pixel_sepia = pixel_sepia.resize((100, 100), Image.NEAREST)
display(scaled_pixel_sepia)
```

## 4. Modification de tous les pixels de l'image <a class="anchor" id="section4"></a>

<ul>
<li> On écrit une fonction grey_transform(img) qui prend en entrée une image et qui donne en sortie l'image transformée en gris </li> <br/>
<li> On écrit une fonction sepia_transform(img) qui prend en entrée une image et qui donne en sortie l'image transformée en sepia</li>
</ul>

```python
def grey_transform(img):
    pixels_sepia = np.zeros((img.size[0],img.size[1],3))
    img_unif1_sepia = Image.new("RGB",img.size)
    for i in range(225):
        for j in range(225): 
            r,g,b = img.getpixel((i,j))
            pixels_sepia[i,j] = np.dot(G,np.array([r,g,b]))
            img_unif1_sepia.putpixel((i,j),(pixels_sepia[i,j,0].astype(int),pixels_sepia[i,j,1].astype(int),pixels_sepia[i,j,2].astype(int)))
    return img_unif1_sepia

def sepia_transform(img):
    pixels_sepia = np.zeros((img.size[0],img.size[1],3))
    img_unif1_sepia = Image.new("RGB",img.size)
    for i in range(225):
        for j in range(225): 
            r,g,b = img.getpixel((i,j))
            pixels_sepia[i,j] = np.dot(S,np.array([r,g,b]))
            img_unif1_sepia.putpixel((i,j),(pixels_sepia[i,j,0].astype(int),pixels_sepia[i,j,1].astype(int),pixels_sepia[i,j,2].astype(int)))
    return img_unif1_sepia
```

```python
# On transforme tous les pixels de l'image en gris et on affiche l'image
img_gris = grey_transform(img)
display(img_gris)
```

```python
# On transforme tous les pixels de l'image en sepia et on affiche l'image
img_sepia = sepia_transform(img)
display(img_sepia)
```

## 5. Rang des matrices $G$ et $S$ et traitement d'image <a class="anchor" id="section5"></a>
Nous remarquons que les matrices de transformation des couleurs ne sont pas inversibles. La matrice $G$ de transformation en gris est de rang $1$ et la matrice $G$ de transformation en Sepia est de rang $2$. Cela signifie qu'en transformant l'image par $G$ ou $S$ on perd de l'information puisqu'un pixel qui s'écrivait avant comme une combinaison linéaire de r, g et b ne s'écrit après transformation en gris que comme combinaison linéaire d'un seul vecteur. On passe d'une dimension $3$ à une dimension $1$ : nous sommes donc en train de projeter l'information. 

De même pour la transformation Sepia, l'image dont les couleurs sont au départ des vecteurs vivant dans $\mathbb{R}^3$ se voit projetés sur une image sepia dont les couleurs sont des vecteurs vivant sur un plan : un espace de dimension $2$. Il y a une perte d'information. 

Pour rendre cela visible, nous allons créer des images qui après transformation en gris ou sepia deviennent uniformes (toute l'information disparaît). 

Pour se faire, nous avons besoin répondre à la question suivante : étant donné deux pixels $p_1$ et $p_2$ à quels conditions sont-ils envoyés sur le même pixel gris ou sur le même pixel sepia? Formalisons. Cela donne : 

$$ \text{soient } p_1,p_2 \in \mathbb{R}^3 \text{ tels que }\, \,  Gp_1 = Gp_2 $$
Alors on a : 
$$ Gp_1 - Gp_2 = 0 $$
$$\iff  G(p_1 - p_2) = 0 $$
$$ \iff p_1 - p_2 \in \mathrm{ker}(G) $$

Ainsi, il faut (et il suffit) que $p_1-p_2$ soit dans le noyau de la matrice $G$ !

Il faut donc déterminer le noyau de $G$. Nous traiterons le cas de la matrice $S$ plus tard. 

```python
# La librairie linalg permet de faire de l'algèbre linéaire. 
# Les opérations suivantes utilisent la diagonalisation de matrices, une notion au programme de l'année prochaine. 
eig = np.linalg.eig(G)

# on note (v1,v2) une base du noyau de G.
v1 = (100*eig.eigenvectors[:,1]).astype(int)
v2 = (100*eig.eigenvectors[:,2]).astype(int)

print("Le vecteur v1 est : " , v1)
print("Le vecteur v2 est : ", v2)

print(np.matmul(G,v1))
print(np.matmul(G,v2))
```

### 5.1 Noyau de $G$<a class="anchor" id="section51"></a>

Nous venons de trouver 
$$ \ker(G) = \mathrm{Vect}(v_1,v_2) $$
avec $v_1 = \begin{pmatrix} -73 \\ -15 \\ 66 \end{pmatrix}$ et $v_2 = \begin{pmatrix} -13 \\ -5 \\ 98 \end{pmatrix}$. Remarquez que la base trouvée pour le noyau de $G$ contient des vecteurs de $\mathbb{R}^3$ qui ne peuvent pas être interpretés comme des pixels puisque les coordonnées ne sont pas dans $[0,255]$. Cela n'est pas tellement important puisque pour la suite ce qui sera important c'est que la différence de deux pixels $p1$ et $p2$ soit dans le noyau de $G$ !

### 5.2 Créer une image qui devient uniforme après transformation en gris. <a class="anchor" id="section52"></a> 

Pour cela, nous allons transformer l'image de départ (solo.jpeg) pour qu'elle ne contienne que des pixels qui diffèrent entre eux d'un élément du noyau de $G$. Comme les valeurs r,g,b des pixels doivent être comprises entre 0 et 255 on va obtenir un nombre fini de pixel possible. Concrètement, on part d'un pixel $p$ quelconque et on construit

$$ p_1 = p + v_1 $$
$$ p_2 = p + 2v_1 $$ 
$$ p_3 = p + 3v_1 $$
$$ q_1 = p + v_2 $$
$$ q_2 = p + 2v_2 $$ 
$$ q_3 = p + 3v_2 $$
$$ r_1 = p + v_1 + v_2 $$

Tous ces pixels ont la même image par la matrice $G$ !
$$ Gp = Gp_1 = Gp_2 = \cdot \cdot \cdot = Gr_1 $$

On va ensuite reconstruire l'image en utilisant uniquement les pixels $(p,p_1,p_2,p_3,q_1,q_2,q_3,r_1)$. 
On utilisera pour cela la fonction closest_pixel qui envoie un pixel de couleur quelconque sur le pixel le plus proche dans une liste donnée en argument. 


```python
def distance(p1, p2):
    return np.linalg.norm(p1-p2)
def closest_pixel(p, pixels):
    min_distance = 1000000
    select_pixel = np.array([0,0,0])
    for i in range(np.shape(pixels)[0]):
        if (distance(p, pixels[i]) < min_distance):
            min_distance = distance(p, pixels[i])
            select_pixel = pixels[i]
    return select_pixel
```

```python
# definition des pixels p1, p2, p3, q1, q2, q3, r1
p = np.array([255,200,0])
p1 = p + v1
p2 = p + 2*v1
p3 = p + 3*v1
q1 = p + v2
q2 = p + 2*v2
q3 = p + 3*v2
r1 = p + v1 + v2
r2 = p + v1 + 2*v2
r3 = p+2*v1 + v2
print("Vérifiez que dans les valeurs des pixels ci-dessous, aucune ne sort de l'intervalle [0,255]")
print(" p : ", p , "\n p1 : ", p1 , " p2 : " , p2 , " p3 : ", p3 , "\n q1 : " , q1 , " q2 : ", q2 , " q3 : " , q3 , "\n r1 : ", r1, " r2 : ", r2, " r3 : " , r3)

pixel_noyauG_2 = Image.new("RGB",(1,1),(v2[0],v2[1],v2[2]))

# Affichage 
print("Couleurs des deux pixels dans le noyau de G")
display(Image.new("RGB",(1,1),(p[0],p[1],p[2])).resize((100, 100), Image.NEAREST))
display(Image.new("RGB",(1,1),(p1[0],p1[1],p1[2])).resize((100, 100), Image.NEAREST))
display(Image.new("RGB",(1,1),(p2[0],p2[1],p2[2])).resize((100, 100), Image.NEAREST))
display(Image.new("RGB",(1,1),(p3[0],p3[1],p3[2])).resize((100, 100), Image.NEAREST))
display(Image.new("RGB",(1,1),(q1[0],q1[1],q1[2])).resize((100, 100), Image.NEAREST))
display(Image.new("RGB",(1,1),(q2[0],q2[1],q2[2])).resize((100, 100), Image.NEAREST))
display(Image.new("RGB",(1,1),(q3[0],q3[1],q3[2])).resize((100, 100), Image.NEAREST))
display(Image.new("RGB",(1,1),(r1[0],r1[1],r1[2])).resize((100, 100), Image.NEAREST))
display(Image.new("RGB",(1,1),(r2[0],r2[1],r2[2])).resize((100, 100), Image.NEAREST))
display(Image.new("RGB",(1,1),(r3[0],r3[1],r3[2])).resize((100, 100), Image.NEAREST))
```

```python
# Transformation de l'image de départ pour obtenir une image qui deviendra uniforme après transformation en gris

# Creation d'une image basée sur solo.jpeg mais ne contenant que les pixels p,p1,...,r3 ci-dessus
img_unif_grey = Image.new("RGB",(225,225))
for i in range(225):
    for j in range(225):
        r,g,b = img.getpixel((i,j))
        pix = np.array([r,g,b])
        closest_pix = closest_pixel(pix,np.array([p,p1,p2,p3,q1,q2,q3,r1,r2,r3]))
        img_unif_grey.putpixel((i,j),(closest_pix[0],closest_pix[1],closest_pix[2]))
        
# Affichage de l'image avant transformation en gris
print(" Image avant transformation en gris")
display(img_unif_grey)

# Affichage de l'image après transformation en gris
print("Image après transformation en gris")
display(grey_transform(img_unif_grey))
```

### 5.3 Même travail avec les images Sepia<a class="anchor" id="section53"></a>

Contrairement à $G$, la matrice $S$ n'est pas de rang $1$ mais de rang $2$. Son noyau est donc de dimension 1. On le calcule et on trouve : 
$$ \ker(S) = \mathrm{Vect}\left(\begin{pmatrix} -36 \\ -3 \\ 92 \end{pmatrix} \right)$$
On note $w = \begin{pmatrix} -36 \\ -3 \\ 92 \end{pmatrix} $

Pour construire comme précédemment une image qui devient unfiorme après transformation en sepia nous possédons moins de possibilités qu'avec la transformation en gris puisqu'il faut créer à partir d'un pixel $p$ des pixels obtenus comme $p + kw$ avec $k$ un entier choisi de tel sorte que les valeurs des pixels restent dans l'intervalle $[0,255]$. Pour maximiser le nombre de pixel nous partons de $p = \begin{pmatrix} 225 \\ 100 \\ 15 \end{pmatrix}$ et nous allons nous autoriser des erreurs d'arrondi en choisissant des $k$ non-entiers. Cela nous permet d'obtenir plus de nuances dans notre image. 
Nous construisons donc :  
$$ p_1 = p + 0.5w $$ 
$$ p_2 = p+w$$
$$ p_3 = p + 1.5w $$ 
$$ p_4 = p+2w$$
$$ p_5 = p+2.5w$$

Comme tout à l'heure on a 
$$ Sp = Sp_1 = Sp_2 = \cdot \cdot \cdot = Sp_5 $$

```python
p = np.array([225, 100, 15])
w = np.array([-37,-4,93])
p1 = (p + 0.5*w).astype(int)
p2 = (p + w).astype(int)
p3 = (p + 1.5*w).astype(int)
p4 = (p + 2*w).astype(int)
p5 = (p + 2.5*w).astype(int)
print("Vérifiez que dans les valeurs des pixels ci-dessous, aucune ne sort de l'intervalle [0,255]")
print(" p : ", p , "\n p1 : ", p1 , " p2 : " , p2 , " p3 : ", p3 , " p4 : " , p4, " p5 : " , p5)
```

```python
# Transformation de l'image de départ pour obtenir une image qui deviendra uniforme après transformation en gris

# Creation d'une image basée sur solo.jpeg mais ne contenant que les pixels p,p1,...,r3 ci-dessus
img_unif_sepia = Image.new("RGB",(225,225))
for i in range(225):
    for j in range(225):
        r,g,b = img.getpixel((i,j))
        pix = np.array([r,g,b])
        closest_pix = closest_pixel(pix,np.array([p,p1,p2,p3,p4]))
        img_unif_sepia.putpixel((i,j),(closest_pix[0],closest_pix[1],closest_pix[2]))
        
# Affichage de l'image avant transformation en gris
print(" Image avant transformation en sepia")
display(img_unif_sepia)

# Affichage de l'image après transformation en gris
print("Image après transformation en sepia")
display(sepia_transform(img_unif_sepia))
```
